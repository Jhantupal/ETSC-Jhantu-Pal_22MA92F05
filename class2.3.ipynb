Ex: Spring-mass-damper with direct integration

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

tstart=0.0
tend=60.0

incre=0.1
t=np.arange(tstart,tend+1,incre)

# initial condition
a=0
b=0.1
x_init=[a,b]

def smdiff(x,t):
  c=4
  k=2
  m=20
  w=2
  F=5*np.sin(w*t)


  dx1dt=x[1]
  dx2dt=(F-c*x[1]-k*x[0])/m

  dxdt =[dx1dt,dx2dt]
  return dxdt

x=odeint(smdiff,x_init,t)
x1=x[:,0]
x2=x[:,1]

plt.plot(t,x1)
plt.plot(t,x2)

plt.xlabel('t')
plt.ylabel('..')
plt.grid()
plt.show()
```
